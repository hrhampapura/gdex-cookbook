## Overview

NSF NCAR's **Geoscience Data Exchange (GDEX)** is the long-term, trusted home for geoscience data at the **National Center for Atmospheric Research (NCAR)**, funded by the **National Science Foundation (NSF)**. GDEX bridges the gap between data producers and the broader scientific community by providing curated, openly accessible datasets alongside the tools and services that can help enable scientific insight.

Whether you are looking to download terabytes of reanalysis data, stream a climate model output directly into a Python workflow, or archive your own research dataset with a citable DOI, GDEX is built to support the full lifecycle of geoscience data.

Visit GDEX at [gdex.ucar.edu](https://gdex.ucar.edu)

## Mission and Purpose

GDEX exists to  make geoscience data **findable, accessible, interoperable, and reusable (FAIR)** to the NCAR scientists and the broader community.

### Open Science

GDEX is committed to open science. Currently, all datasets are freely accessible to anyone, anywhere in the world without institutional or technical barriers. Additionally, data services are continuously being developed and expanded to minimize friction with obtaining and interacting with data. 

### Supporting the Full Data Lifecycle

GDEX supports data from initial archival through active use in research workflows:

- **Ingest & Curation**: Data submitted to GDEX is reviewed, documented, and formatted to community standards
- **Discovery**: Rich metadata and search tools help users find the right dataset quickly
- **Access**: Multiple access pathways — from bulk download to cloud streaming
- **Citation**: Persistent identifiers (DOIs) ensure datasets are citable and contributions are credited
- **Preservation**: GDEX provides long-term stewardship and data backups
- **Stewardship**: Science-educated staff assist users in solving problems and forwarding science.


:::{note}
All GDEX datasets are freely accessible — no institutional login is required to browse, search, or download public datasets.
:::

## Services and Tools

GDEX offers a comprehensive suite of services to help researchers at every stage of their workflow.

### Discovery and Search

The [GDEX web portal](https://gdex.ucar.edu) provides powerful tools for finding data:

- **Full-text and faceted search** across thousands of datasets with filtering by variable, domain, date range, and format
- **Rich metadata** including dataset documentation, version history, and related publications
- **Dataset previews** and quick-look visualizations to assess data before downloading

### Data Access Pathways

GDEX meets researchers wherever they work:

| Pathway | Best For |
|---|---|
| **Web download** | Browsing and ad-hoc downloads via the portal |
| **Globus transfer** | Bulk, reliable transfer of large datasets to HPC systems |
| **OPeNDAP/HTTPS** | Programmatic and subsetting access from scripts |
| **Cloud streaming (Zarr/Kerchunk)** | Analysis-ready access without downloading |
| **Python/HTTPS APIs** | Integration directly into data science workflows |

### Subsetting and Format Conversion

Many datasets support server-side subsetting, allowing you to request only the variables, spatial domain, and time range you need — reducing download size and time significantly.

### Compute-Proximate Access

For large datasets where moving data is impractical, GDEX data is accessible from NSF NCAR computing resources including the **Derecho** supercomputer and the **Casper** analysis cluster. Researchers with allocations can run analyses directly alongside the data.

### Metadata and Documentation

Each dataset includes:

- Dataset-level metadata including abstract, variables, resources, and citations
- File-level metadata following community standards and uses [CF Conventions](http://cfconventions.org/) when possible
- Documentation and software files when applicable
- Version tracking so analyses can be reproduced precisely
- Links to related publications and data sources

## Data Submission

GDEX accepts data submissions from the broader scientific community. Archiving your data with GDEX is a way to ensure it reaches the widest possible audience and remains accessible for the long term.

### Who Can Submit?

Data submission is open to:

- **NSF NCAR scientists and collaborators** conducting funded research
- **University researchers** and UCAR member institutions
- **Project teams** with NSF or other federally funded grants that require a data management plan

If you are unsure whether your dataset is a good fit, contact the GDEX team at [datahelp@ucar.edu](mailto:datahelp@ucar.edu).

### What Types of Data Are Accepted?

GDEX prioritizes datasets that:

- Are of **broad community interest** for atmospheric, oceanic, or related geoscience research
- Come with sufficient **documentation and metadata** to make the data reusable
- Are in **community-standard formats** (or can be converted to them)
- Represent **unique or irreplaceable** observational records or model output

### The Submission Process

1. **Contact the GDEX team** — [Submit](https://gdex.ucar.edu/resources/submit-data/) your data and discuss its suitability with the GDEX team
2. **Prepare your data** — Work with the GDEX team to ensure your data and metadata meet community standards
3. **Transfer your data** — GDEX supports transfer via Globus, HTTPS, or other methods depending on data volume
4. **Curation and review** — The GDEX team reviews, documents, and organizes your dataset
5. **Publication and DOI** — Your dataset is published with a persistent DOI, making it citable in publications

### Benefits of Archiving with GDEX

- Your dataset receives a **citable DOI** that can be included in publications
- Long-term preservation and backups by a trusted, **CoreTrustSeal certified** repository
- Broad discoverability through GDEX search and external indexes (e.g., Google Dataset Search)
- Compliance with **NSF and journal data sharing requirements**
- Access to a variety of value added **services**
- Your data becomes part of a collection used by researchers worldwide


:::{tip}
Contact [datahelp@ucar.edu](mailto:datahelp@ucar.edu) before investing significant effort in data preparation. The GDEX team can confirm your dataset is a good fit and advise on format and metadata requirements.
:::

## How GDEX Enables Science

GDEX is more than a file repository — it is infrastructure that enables scientific discovery at scale.

### Reproducible Research

By providing stable, versioned, citable datasets, GDEX makes it possible to reproduce published analyses years after the fact. When a paper cites a GDEX dataset DOI, any researcher in the world can access the exact same data used in that study.

### Lowering Barriers to Large-Scale Analysis

Some of the most scientifically important datasets are also the largest. ERA5 global reanalysis alone spans petabytes. GDEX removes the logistical barriers to working with these datasets by:

- Hosting data on infrastructure optimized for high-throughput access
- Providing cloud-optimized formats (Zarr, Kerchunk) that enable streaming analysis without full downloads
- Supporting compute-proximate workflows on NSF NCAR HPC systems

### Enabling Multi-Dataset Science

Because GDEX curates data to common standards, it becomes easier to combine datasets — for example, comparing reanalysis products or merging model output with observations. Consistent metadata and file formats reduce the data wrangling burden and let researchers focus on science.

### Supporting Data-Driven Discovery

The scale and diversity of GDEX collections make them well-suited for emerging methods in geoscience:

- **Machine learning** applications that require large training datasets
- **Long-term trend analysis** using multi-decadal observational records
- **Model evaluation** comparing simulations against observational benchmarks
- **Climate impact studies** combining atmospheric, oceanic, and land surface data

### Example: Streaming Data Directly into Python

Cloud-optimized formats hosted by GDEX allow researchers to open petabyte-scale datasets with a single line of Python — no full download required:

```python
import xarray as xr

# Open a cloud-optimized GDEX dataset
ds = xr.open_zarr("https://data.gdex.ucar.edu/d633000/e5.oper.an.sfc.zarr/e5.oper.an.sfc.100u.zarr", consolidated=True)
ds["VAR_100U"].sel(time="2020-01-01").plot()
```

This workflow is possible because GDEX invests in cloud-optimized formats and stable, high-bandwidth access infrastructure.

## Getting Help

The GDEX team provides direct support for data access, discovery, and submission questions.

### Data Help Desk

The primary support channel is the **GDEX Data Help Desk**:

- **Email**: [datahelp@ucar.edu](mailto:datahelp@ucar.edu)
- **Response time**: typically 1–2 business days
- **Best for**: data access problems, subsetting questions, format guidance, submission inquiries, and anything not covered by documentation

When emailing, include the dataset identifier (e.g., `ds083.2`) and a description of what you were trying to do — this significantly speeds up the response.

### Common Questions

| Question | Where to look first |
|---|---|
| I can't find a dataset | [Dataset search](https://gdex.ucar.edu/gsearch/dataset-search/) — try broader terms or filtering by variable |
| I need access to a restricted dataset | Check the dataset page for access instructions, or email the help desk |
| My download failed or is very slow | Try [Globus](https://www.globus.org/) for large transfers, or email the help desk |
| I want to submit data to GDEX | See the [submission page](https://gdex.ucar.edu/resources/submit-data/) or the Data Submission notebook in this cookbook |
| I need help with a specific Python workflow | Browse the service notebooks in this cookbook |
| A dataset seems incorrect or has a quality issue | Email the help desk with the dataset ID and a description of the issue |

## Overview

GDEX accepts data submissions from the broader scientific community. Archiving your dataset with GDEX
gives it a citable DOI, long-term preservation by a CoreTrustSeal-certified repository, and
broad discoverability through the GDEX portal and external indexes such as Google Dataset Search.

This notebook walks through what to expect at each stage of the submission process, from initial
contact through to publication.

## Who Can Submit?

Data submission is open to a wide range of contributors:

- **External researchers** — anyone with a free GDEX account can submit data via the online form
- **NSF NCAR scientists and collaborators** — staff and project collaborators conducting funded research
- **University researchers and UCAR member institutions** — particularly those with NSF or other
  federally funded grants that include a data management plan
- **Project teams** — multi-institution projects producing community datasets of broad interest

If you are unsure whether your dataset is a good fit, reach out before investing significant
effort in preparation.

:::{tip}
Email [datahelp@ucar.edu](mailto:datahelp@ucar.edu) early — the GDEX team can quickly assess
whether your dataset meets the criteria and advise on format and metadata requirements before
you begin formal submission.
:::


## Submission Pathways

There are two routes into GDEX depending on your role:

### Submit Data — All Users

Any researcher can submit a dataset directly using the online form at
[gdex.ucar.edu/resources/submit-data-request](https://gdex.ucar.edu/resources/submit-data-request/).
You share basic details about the dataset; the GDEX team reviews the submission and responds by email.

- **Review time**: approximately 30 minutes for NCAR employees; up to one week for external submissions
- **Outcome**: email notification of approval or rejection, with guidance on next steps if approved

### Suggest a Dataset — NCAR Employees Only

NCAR staff can suggest datasets for inclusion in the GDEX catalog using a separate internal form at
[gdex.ucar.edu/resources/suggest-dataset](https://gdex.ucar.edu/resources/suggest-dataset/).
These go through an internal review for GDEX compatibility and curation requirements.

:::{note}
The **Suggest Dataset** pathway is restricted to NCAR employees. External researchers and collaborators
should use the **Submit Data** form instead.
:::


## What Makes a Dataset a Good Fit?

The GDEX team evaluates submissions on the following criteria:

| Criterion | Details |
| --- | --- |
| **Community interest** | Broad relevance to atmospheric, oceanic, or related geoscience research |
| **Documentation** | Sufficient metadata and documentation to make the data reusable by others |
| **Format** | Community-standard formats (NetCDF4, GRIB2, HDF5) or convertible to them |
| **Uniqueness** | Irreplaceable observational records, unique model output, or significant value-add over existing holdings |
| **Compliance** | Dataset must meet [GDEX Terms & Conditions](https://gdex.ucar.edu/resources/docs/gdex-terms-conditions/) |

Datasets do not need to be large — niche but well-documented records of lasting scientific value are
welcome alongside petabyte-scale model archives.


## Preparing Your Dataset

Before submitting, work through this checklist to give your dataset the best chance of approval
and to minimize back-and-forth with the GDEX team:

**Data files**
- [ ] Files are in a community-standard format (NetCDF4 preferred; GRIB2, HDF5, or CSV where appropriate)
- [ ] Variable names and units follow [CF Conventions](http://cfconventions.org/) where applicable
- [ ] Files are complete, with no placeholders or partial records

**Metadata and documentation**
- [ ] Dataset abstract describing what the data is, how it was produced, and its scientific purpose
- [ ] Variable-level descriptions including units, fill values, and any processing notes
- [ ] Spatial and temporal coverage clearly defined
- [ ] Contact information for the dataset PI
- [ ] Links to related publications or grant numbers

**Rights and licensing**
- [ ] You have the right to archive and share the dataset
- [ ] Any third-party data included is properly licensed for redistribution
- [ ] Preferred license identified (GDEX defaults to open access)

:::{important}
Datasets that arrive without adequate documentation are the most common reason for submission delays.
A complete abstract and variable descriptions dramatically speed up the curation process.
:::


## The Submission Process

Once you submit the online form, the process unfolds in roughly five stages:

1. **Initial review** — GDEX staff assess the dataset for scope, format, and documentation fit.
   You receive an email with approval or rejection (and the reason, if rejected).

2. **Data transfer** — Approved submissions move to transfer. GDEX supports Globus, HTTPS, and
   direct transfer for large volumes. The team will advise on the best method for your dataset size.

3. **Curation** — GDEX staff review and organize the files, validate metadata, and apply
   any format conversions needed to meet community standards.

4. **DOI assignment** — The dataset receives a persistent DOI, making it citable in publications
   and discoverable through external indexes.

5. **Publication** — The dataset goes live on the GDEX portal with its landing page, metadata
   record, and access links.

:::{tip}
After your dataset is published, add the DOI to your papers, grant reports, and data management
plan. Citations are tracked and contribute to demonstrating the impact of your work.
:::

## Summary

Submitting data to GDEX involves three key steps: filling out the online submission form,
working with the GDEX team to transfer and curate your files, and receiving a citable DOI
once the dataset is published. Datasets of any size are considered — what matters most is
scientific value, adequate documentation, and a community-standard format.

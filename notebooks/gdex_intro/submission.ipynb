# Submitting a Data Request to GDEX

---

## Overview

GDEX allows users to submit custom subset requests — specifying spatial region, time range, and variables — and then download the resulting files. This notebook walks through the full request lifecycle.

1. Obtain an authentication token from your GDEX profile
2. Retrieve a control file template for a dataset
3. Submit a subset request
4. Poll the request status until complete
5. Retrieve and download the output files

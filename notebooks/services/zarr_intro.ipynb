## Overview

GDEX provides Analysis-Ready Cloud-Optimized (ARCO) versions of select datasets in Zarr or virtualized Zarr (kerchunk, virtualizarr) format. ARCO data is chunked and indexed for fast random access — ideal for large-scale analysis.

1. Check whether a dataset has ARCO data available
2. Browse and search ARCO variables
3. Open an ARCO Zarr store with xarray

## API Conventions

All endpoints live under `https://gdex.ucar.edu/api/`. Dataset IDs use the format `d######` (six digits). Every successful response follows this envelope returned in JSON format:

```json
{
  "status": "ok",
  "http_response": 200,
  "error_messages": [],
  "contact": "datahelp@ucar.edu",
  "data": { ... }
}
```

````python
BASE_URL = "https://gdex.ucar.edu"
````

## Authentication

Most read endpoints are public and do **NOT** require authentication. However, Subset and purge endpoints require authentication via:
- **Token**: `?token=<your_token>` — get yours at `https://gdex.ucar.edu/accounts/profile/`


## Your First API Call

`GET /api/datasets/` lists all available datasets with their IDs and titles.

````python
# list all datasets
all_datasets_url = f"{BASE_URL}/api/datasets/"
resp = requests.get(all_datasets_url)
envelope = resp.json()
print(envelope["data"]["count"], "datasets")
for ds in envelope["data"]["datasets"][:5]:
    print(ds["id"], "—", ds["title"])
````

---

## Summary

The GDEX API follows a consistent REST pattern: `GET https://gdex.ucar.edu/api/<endpoint>/`. Public endpoints need no auth; subset and purge endpoints require a token from your GDEX profile.

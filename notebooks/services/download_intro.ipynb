## The GDEX web UI interface
.  
This section explains how to navigate from the GDEX landing page to a dataset’s **Download** page, then use the web interface to discover files efficiently. You will learn where to find core controls such as file path browsing, search, sorting, and filter options.

The portal is useful for both quick manual downloads and preparing larger scripted workflows. In particular, it helps you:

- narrow file lists by available metadata filters (for example, date range, filename pattern, or directory path),
- inspect some meta data details before downloading,
- copy direct file links for HTTP-based automation.

### Landing page

The [GDEX web portal](https://gdex.ucar.edu/) is a centralized place to access all different kinds of services provided by GDEX. To begin your search, you can directly perform a syntax search for the dataset you are interested in.
:::{figure} ../../images/gdex-web-landing.png
:alt: GDEX landing page
:width: 90%

Landing page of the GDEX web portal to get started on the dataset search
:::

### Dataset Page

The dataset page provides an overview of the selected dataset, including:

- **Dataset metadata**: Title, description, and temporal coverage  
- **Access methods**: Available download options (HTTP, Globus, POSIX)  
- **Citation**: How to cite the dataset  
- **Dataset metrics**: Statistics on dataset usage  

:::{figure} ../../images/gdex-web-dataset.png
:alt: dataset landing page
:width: 90%

Dataset page displaying dataset-specific information and available options
:::


### Data Access Tab

The Data Access tab at the top of the dataset page displays all available data access methods provided by GDEX. To download files to your local machine, locate the "Web Download (HTTPS)" column on the left. Files are typically organized into groups for easier navigation, though an "All Data Products" view is available to browse the entire dataset.

:::{figure} ../../images/gdex-web-dataaccess.png
:alt: data access tab for every dataset
:width: 90%

Find all available data access options for this specific dataset
:::


### Dataset Group/Subgroup

Because datasets can contain many file types and large file collections, GDEX may provide **group/subgroup** options to make navigation faster and easier.

Groups and subgroups can be organized by:

- **Temporal frequency** (for example, daily, monthly, hourly)
- **Simulation domain** (for example, atmosphere, ocean, land)
- Other dataset-specific categories that improve search efficiency

:::{figure} ../../images/gdex-web-group.png
:alt: group/subgroup view of the dataset
:width: 90%

In this example, we select the **ocean monthly** group to focus on ocean-domain simulation output at monthly frequency.
:::


### Dataset download

Now we finally get to the data file level. In this interface, you can further narrow down the file list by applying filters. By entering a filter keyword, you can reduce the file list to easily select the files you need to download.

:::{figure} ../../images/gdex-web-filter.png
:alt: data file download page 
:width: 90%

After entering the filter keyword, you can select all the files needed and generate a download script (csh, python, or Jupyter Notebook) for batch downloads. Alternatively, you can click on individual filenames on the web page to instantly download a single data file.
:::

### Advanced search

For some datasets, if sufficient metadata was captured during archival, the **Web Download (HTTPS)** interface provides two options:

- **File List**
- **File Search**

:::{figure} ../../images/gdex-web-filesearch.png
:alt: file search and file list options
:width: 90%

For datasets with rich metadata, both **File List** and **File Search** are available.
:::

**File List** provides the standard hierarchical view of groups and files described earlier.  
**File Search** offers an advanced filtering interface that allows detailed selection by metadata attributes, such as variable, time range, and available temporal frequency.

:::{figure} ../../images/gdex-web-metafilter.png
:alt: detail file search based on meta data
:width: 90%

**File Search** enables advanced filtering to narrow file selection using dataset-provided metadata.
:::

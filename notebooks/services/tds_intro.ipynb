## Accessing TDS via the Web Browser

The GDEX TDS catalog can be explored interactively through any web browser — no Python or programming required. This section walks through the key steps.

### Step 1: Open the GDEX TDS Catalog

Navigate to the GDEX TDS catalog root in your browser:

```
https://tds.gdex.ucar.edu/thredds/catalog/catalog.html
```

You will see a list of top-level dataset collections. Each entry corresponds to a GDEX dataset identified by its dataset ID (e.g., `d010049`).

:::{figure} ../../images/gdex-tds-landing.png
:alt: GDEX TDS catalog root page showing dataset collections
:width: 90%

GDEX TDS catalog root — list of available dataset collections.
:::

### Step 2: Browse to a Dataset

Click a dataset ID folder (e.g., `d010049`) to enter its sub-catalog. Sub-catalogs are organized by year or other time partitions. Click through until you reach a directory containing individual files.

:::{figure} ../../images/gdex-tds-subcattofile.png
:alt: Click through until you see individual files
:width: 90%

Navigating into a dataset sub-catalog/files.
:::

### Step 3: View Available Services for a File

Click on an individual file name to open its dataset page. The page lists all access services (OPeNDAP, NCSS, WMS, HTTP download, etc.) with direct links — you can copy any URL for use in Python or paste it into a browser to download.

## The GDEX TDS Catalog

The GDEX TDS catalog root URL is `https://tds.gdex.ucar.edu/thredds/catalog/catalog.html`. You can navigate sub-catalogs by dataset ID (dnnnnnn where n are digits) . Here, we demostrate a dataset d010049 : Global Ocean Physics Reanalysis as a example below.

````python
# open the GDEX TDS catalog
cat = TDSCatalog("https://tds.gdex.ucar.edu/thredds/catalog/files/d010049/1993/catalog.xml")
print(cat.datasets)
````

## Available Data Services

TDS provides multiple access protocols for each dataset. Common ones include:
- **OPeNDAP** — remote array subsetting, works with xarray
- **NCSS** — NetCDF Subset Service, time/spatial subsetting
- **WMS** — Web Map Service for visualization

````python
cat.datasets.values()
````

````python
list(cat.datasets.values())[0]
````

````python
# inspect services available for a dataset
ds = list(cat.datasets.values())[0] # we are getting the first dataset in the catalog

print(f'Services available for this dataset ({ds.name}):')
for service_name, service_url in ds.access_urls.items():
    print(f'- {service_name}: {service_url}')
# print(ds.access_urls)
````

### Services available for this dataset

The dataset `mercatorglorys12v1_gl12_mean_19930101_R19930106.nc` exposes multiple access protocols through the GDEX THREDDS Data Server.

- **OpenDAP** — Remote data access and subsetting over HTTP; commonly used with tools like `xarray` for analysis without downloading full files.
- **NetcdfSubset** — NetCDF Subset Service (NCSS) for requesting specific variables, time ranges, vertical levels, and geographic regions.
- **CdmRemote** — Binary remote access protocol for CDM-compatible clients; useful for efficient programmatic data retrieval.
- **CdmrFeature** — Feature-oriented CDM remote access for structured scientific data types (for example, grids and point features).
- **JupyterNotebook** — Launch/open notebook-based workflows connected to the dataset in an interactive environment.
- **WCS** — Web Coverage Service for retrieving geospatial raster/coverage data with spatial and dimensional subsetting.
- **WMS** — Web Map Service for requesting rendered map images of dataset layers for visualization.
- **HTTPServer** — Direct file download access over HTTP.
- **ISO** — ISO-standard metadata representation for cataloging and discovery.
- **NCML** — NetCDF Markup Language view describing dataset structure and metadata in XML form.
- **UDDC** — Unified Data Discovery Convention metadata endpoint for interoperable dataset discovery.

These services support different workflows, from quick map visualization (**WMS**) to analysis-ready subset extraction (**NetcdfSubset/OpenDAP**) and metadata discovery (**ISO/NCML/UDDC**).

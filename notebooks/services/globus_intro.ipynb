# Transferring GDEX Data with Globus

---

## Overview

Globus enables fast, reliable transfers of large datasets from GDEX to your HPC system (or to a personal computer using [Globus Connect Personal](https://www.globus.org/globus-connect-personal)).

With a completed GDEX subset request, you can start a transfer directly through the GDEX API. See the [GDEX API Globus documentation](https://api.gdex.ucar.edu/documentation/#/globus) for endpoint details.

In this notebook, you will:

1. Understand what a Globus endpoint is
2. Prepare a Globus account and destination endpoint
3. Initiate a transfer using the Globus web UI
4. Monitor transfer status in the Globus web UI

## Globus End Point
A **Globus endpoint** is a storage location that Globus can read from or write to during a transfer.

For this workflow:

- **Source endpoint**: the GDEX-managed endpoint that contains the GDEX datasets.
- **Destination endpoint**: your endpoint (for example, an HPC system, campus storage, or Globus Connect Personal on your computer).

Each endpoint has a unique identifier called an **Endpoint UUID**.  
<!-- You provide your destination UUID to the GDEX API so Globus knows where to send the files. -->

:::{tip}
- Verify your destination endpoint is visible in [app.globus.org](https://app.globus.org).
- Make sure the endpoint is **active/online** and you have write permission to the target path.
- Keep the UUID exact (including dashes), since it is required in the transfer request.
:::

## Globus Account and Endpoint

You can create a **Globus account** using one of several identity providers, including:

- **ORCID iD**
- **Google**
- **GitHub**
- **Institutional login** (if supported by your organization)

:::{figure} ../../images/globus-login.png
:alt: Globus login page
:width: 90%

Sign in to Globus using any supported identity provider.
:::

After signing in, you can add your personal computer as a destination endpoint by installing **Globus Connect Personal**.

:::{figure} ../../images/globus-personnel.png
:alt: Globus Connect Personal
:width: 90%

Choose your operating system, then download and install Globus Connect Personal on your local machine.
:::

- Installation guide: [https://docs.globus.org/globus-connect-personal/install/](https://docs.globus.org/globus-connect-personal/install/)
- Download page: [https://www.globus.org/globus-connect-personal](https://www.globus.org/globus-connect-personal)

Once Globus Connect Personal is installed and running, your personal endpoint will appear in Globus and can be used as the transfer destination.

:::{figure} ../../images/globus-endpoint.png
:alt: Globus web UI endpoint view
:width: 90%

Search for the name of your personal endpoint in the Globus web app. You can also search for public collections such as **NSF NCAR GDEX Dataset Archive** using the collection search bar.
:::


## Initiate Transfer

Once the source (**NSF NCAR GDEX Dataset Archive**) and destination endpoints are configured in the Globus File Manager, you can proceed with the transfer.

:::{figure} ../../images/globus-manager-local.png
:alt: Globus File Manager showing source and destination endpoints
:width: 90%

In the **File Manager** tab, set the source endpoint (GDEX) on one side and your destination endpoint on the other.
:::

Navigate to the desired folder on the GDEX endpoint and select the files or directories you wish to transfer.

:::{figure} ../../images/globus-transfer.png
:alt: Globus file selection and transfer initiation
:width: 90%

Browse the GDEX endpoint to locate your files, select them, and choose the destination path on your endpoint.
:::

Click the **Start** button to initiate the transfer. The transfer will run in the background — you do not need to keep your browser open. You can monitor progress in the **Activity** panel on the left, and Globus will send you an email notification when the transfer is complete.

## GDEX Globus Service

You can access the **NSF NCAR GDEX Dataset Archive** endpoint in two ways:

1. Search directly in the Globus web UI
2. Use dataset-specific links from the GDEX web portal's data access tab

:::{figure} ../../images/globus-gdex-link.png
:alt: GDEX Globus link in the data access page
:width: 90%

The GDEX web portal provides dataset-level links to the **NSF NCAR GDEX Dataset Archive** endpoint.
:::

Globus transfers are the recommended method for moving large GDEX datasets to HPC systems. The GDEX API can initiate the transfer directly from a completed request index.
